# 데이터 수집
https://www.goingsool.com/
해당 사이트의 양조장 정보를 동적크롤링을 통해 가져온다

```python
# pip install pandas
# !pip install numpy
# !pip install selenium
# !pip install beautifulsoup4
# !pip install -U selenium
# !pip install webdriver-manager
```

```python
# 관련 모듈 import
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()

# 웹사이트 열기
driver.get('https://www.goingsool.com/')
cnt = 0

data = []

for i in range(1,7):
    element_id = f"{i:03}"
    #print(element_id)
    element = driver.find_element(By.ID, element_id)
    element.click()
    time.sleep(1)

    li_elements = driver.find_elements(By.CSS_SELECTOR, ".list li")

    for li in li_elements:
        info_text = li.text.split("\n")
        #print(li.text, "\n")

        if(li.text == "해당 권역에 등록된 양조장이 없습니다."):
            continue


        # 데이터 파싱 로직
        name = info_text[0]
        address = info_text[1].split("주소 ")[1]
        drink_type = info_text[2].split("주종 ")[1]
        reservation = info_text[3].split("예약문의 ")[1]
        minimum_people = info_text[4].split("최소인원 ")[1]

        try:
            image_element = li.find_element(By.TAG_NAME, "img")
            image_src = image_element.get_attribute("src")
        except:
            image_src =  "null"  # 이미지가 없는 경우 예외 처리

        data.append({
            "양조장 이름": name,
            "주소": address,
            "주종": drink_type,
            "예약문의": reservation,
            "최소인원": minimum_people,
            "이미지 URL": image_src
        })

df = pd.DataFrame(data)

# 데이터프레임 출력
print(df)

# 브라우저 종료
driver.quit()
```

```python
print(df)
df.to_csv('./collected_data.csv', index=False, encoding= 'cp949')
```

## 카카오맵 API 연동


```python
import requests, json

def get_location(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    headers = {"Authorization": "KakaoAK cd6e74b7d32c299e048bab9bf7c33fa8"}  # 실제 API 키로 교체해야 합니다.
    response = requests.get(url, headers=headers)
    api_json = response.json()

    if 'documents' in api_json and api_json['documents']:
        address_info = api_json['documents'][0]['address']
        crd = {"lat": float(address_info['y']), "lng": float(address_info['x'])}
        return crd
    else:
        return {"lat": None, "lng": None}

# 위도와 경도 컬럼을 데이터 프레임에 추가
df['위도'] = None
df['경도'] = None

# 각 주소에 대해 위도와 경도를 가져와서 데이터 프레임에 저장
for index, row in df.iterrows():
    crd = get_location(row['주소'])
    df.at[index, '위도'] = crd['lat']
    df.at[index, '경도'] = crd['lng']

# 변경된 데이터 프레임 확인
print(df)

# 변경된 데이터 프레임을 CSV 파일로 저장
df.to_csv('./updated_data.csv', index=False, encoding='cp949')
```
