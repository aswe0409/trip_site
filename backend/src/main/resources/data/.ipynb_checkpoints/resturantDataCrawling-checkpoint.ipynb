```python
# 관련 모듈 import
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
driver.get('https://map.kakao.com/')

search_box = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
search_box.send_keys('음식점')
search_box.send_keys(Keys.RETURN)

# 잠시 대기 (필요한 경우 조정)
time.sleep(1)

import re

def get_store_data():
    stores = driver.find_elements(By.CSS_SELECTOR, 'li.PlaceItem')
    store_data = []

    for store in stores:
        try:
            name = store.find_element(By.CSS_SELECTOR, 'a.link_name').text
            rating_text = store.find_element(By.CSS_SELECTOR, 'span.score').text
            rating = float(re.search(r'\d+(\.\d+)?', rating_text).group())

            if rating >= 3.5:
                address = store.find_element(By.CSS_SELECTOR, 'p.loca > span.addr').text
                store_data.append({'name': name, 'rating': rating, 'address': address})
        except Exception as e:
            print(f"Error: {e}")

    return store_data

# 첫 페이지 데이터 수집
all_store_data = get_store_data()

# 다음 페이지 버튼을 찾아서 클릭하고 데이터 수집 (원하는 만큼 반복)
for _ in range(5):  # 예: 5페이지까지 크롤링
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, 'a.next')
        next_button.click()
        time.sleep(2)  # 페이지 로딩 대기
        all_store_data.extend(get_store_data())
    except:
        break  # 다음 버튼이 없는 경우

# 수집된 데이터 출력
for store in all_store_data:
    print(store)

# 드라이버 종료
driver.quit()
```
