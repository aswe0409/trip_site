```python
# 관련 모듈 import
import pandas as pd
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
```

```python
# 웹드라이버 경로 설정
driver = webdriver.Chrome()

# 웹사이트 열기
url = "https://korean.visitkorea.or.kr/kfes/list/wntyFstvlList.do"
driver.get(url)

# 스크롤을 하여 모든 li 태그를 로드
scroll_pause_time = 2  # 대기 시간을 늘림
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 페이지 끝까지 스크롤
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

# 모든 li 태그 찾기
ul_tag = driver.find_element(By.ID, 'fstvlList')
li_tags = ul_tag.find_elements(By.TAG_NAME, 'li')

# li_tags의 길이 출력
print(f"Number of li tags: {len(li_tags)}")

# 데이터를 저장할 리스트 초기화
data = []

# 각 li 태그 클릭하여 상세보기 페이지 데이터 추출
for li in li_tags:  # 모든 항목 선택
    # li 태그 클릭
    li.click()
    time.sleep(3)  # 페이지가 로드될 시간을 충분히 줌
    
    # 상세보기 페이지의 데이터 추출
    try:
        # 제목
        try:
            title = driver.find_element(By.ID, 'festival_head').text
        except:
            title = None
        try:
            description = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_info_content > div.m_img_fst > div').text
        except:
            description =None
        
        # 이미지 URL
        try:
            img_url = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.detail_img_box > a > img').get_attribute('src')
        except:
            img_url = None
        # 기간
        try:
            period = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(1) > p').text
            start_date, end_date = period.split(" ~ ")
        except:
            period = None 
        # 주소
        try:
            address = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(2) > p').text
        except:
            address = None
        # 금액
        try:
            price = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(3) > p').text
        except:
            price = None
        # 주최기관
        try:
            organizer = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(4) > p').text
        except:
            organize = None
        # 전화번호
        try:
            phone = driver.find_element(By.CSS_SELECTOR, '#mainTab > div > div > section.poster_detail > div > div.poster_detail_wrap > div > div.img_info_box > ul > li:nth-child(5) > a > p').text
        except:
            phone = None
        # 공식 홈페이지
        try:
            homepage = driver.find_element(By.CLASS_NAME, 'homepage_link_btn').get_attribute('href')
        except:
            homepage = None
        
        # 데이터 리스트에 추가
        data.append([title, img_url, start_date, end_date, address, price, organizer, phone, homepage])
        
    except Exception as e:
        print(f"상세 페이지를 로드하는 데 실패했습니다: {e}")
    
    # 다시 목록 페이지로 돌아가기
    driver.back()
    time.sleep(3)  # 페이지가 로드될 시간을 충분히 줌
    ul_tag = driver.find_element(By.ID, 'fstvlList')
    li_tags = ul_tag.find_elements(By.TAG_NAME, 'li')

# 드라이버 종료
driver.quit()

# 데이터 프레임 생성
df = pd.DataFrame(data, columns=['제목', '이미지 URL', '시작일', '종료일', '주소', '금액', '주최기관', '전화번호', '공식 홈페이지'])
print(df)
```

```python
# 0번부터 시작하는 인덱스 1번 부터
df.index = df.index +1
df
```

```python
df.to_csv('./festival_data.csv', index=False, encoding= 'UTF-8-sig')
```
